```python
import pandas as pd
```

```python
data = pd.read_csv("search_cohort_raw.csv")
```

```python
df = data.groupby(by="user_id")
df.size()
```

```python
date_start = df['date'].min()
date_start.rename('date_start')
```

```python
data_cohort = pd.merge(data,date_start,on='user_id')
data_cohort = data_cohort.rename(columns={"date_x":'date_return','date_y':'date_start'})
```

```python
df2 = data_cohort.groupby(['date_start','date_return'])
final_data = df2.size()
final_data = final_data.to_frame('user_count')
```

```python
final_data
```

```python
cohort_data = final_data['user_count']
cohort_data = cohort_data.unstack(1)
cohort_data
```

```python
totaluser = final_data.groupby(by='date_start')
totaluser = totaluser['user_count'].max()
normalized_table = cohort_data.div(totaluser,axis=0)*100
normalized_table
```
